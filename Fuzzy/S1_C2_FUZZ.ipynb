```python
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import numpy as np
import matplotlib.pyplot as plt
```

Ejercicio 2

```python
temp = ctrl.Antecedent(np.arange(60, 110, 1), 'Temperatura')
temp['Media']=fuzz.trapmf(temp.universe, [60, 60, 70, 80])
temp['Alta']=fuzz.trapmf(temp.universe, [70, 80, 90, 100])
temp['MuyAlta']=fuzz.trapmf(temp.universe, [90, 100, 110,110])
temp.view()
```
